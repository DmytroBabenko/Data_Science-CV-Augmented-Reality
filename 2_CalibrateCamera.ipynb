```python
import cv2 as cv
from matplotlib import pyplot as plt
from skimage import io
from camera_calibrator import CameraCalibrator
from os import listdir
from os.path import isfile, join
```

```python
def imshow(img):
    fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
    plt.axis('off')
    plt.imshow(img)
```

```python
calibrate_imgs_dir = "images/calibration/elipse"
files = [f for f in listdir(calibrate_imgs_dir) if isfile(join(calibrate_imgs_dir, f))]
calibrate_imgs = files[1:] if files[0] == '.DS_Store' else files
calibrate_imgs = [f"{calibrate_imgs_dir}/{img}" for img in calibrate_imgs]
calibrate_imgs
```

```python
calibrator = CameraCalibrator()
```

```python
calibrator.calibrate(calibrate_imgs)
```

```python
distortion_image_file = "images/calibration/Test_Image2.png"
distortion_image = cv.imread(distortion_image_file)
```

```python
imshow(cv.cvtColor(distortion_image, cv.COLOR_BGR2RGB))
```

```python
undistoted_image = calibrator.undistort(distortion_image_file)
```

```python
imshow(cv.cvtColor(undistoted_image, cv.COLOR_BGR2RGB))
```
